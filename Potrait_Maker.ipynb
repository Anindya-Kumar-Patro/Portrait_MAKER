```
from google.colab.patches import cv2_imshow
```

```
import cv2
```

```
img = cv2.imread("anindya.jpeg")
```

```
cv2_imshow(img)
```

```
grayimage = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
invimage = cv2.bitwise_not(grayimage)
```

```
gasblur = cv2.GaussianBlur(invimage , ksize =(21,21), sigmaX=0, sigmaY = 0)
cv2_imshow(gasblur)
```

```
def dodgeV2(x,y):
  return cv2.divide(x,255-y, scale = 256)

potrait = dodgeV2(grayimage, gasblur)
cv2_imshow(potrait)
```

```
imgtarget = "Potrait.jpg"
cv2.imwrite(imgtarget,potrait)
```
